### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda["tienda"] = "Tienda 1"
tienda2["tienda"] = "Tienda 2"
tienda3["tienda"] = "Tienda 3"
tienda4["tienda"] = "Tienda 4"
```

```python
df = pd.concat([tienda, tienda2, tienda3, tienda4])

df.head()
```

#1. Análisis de facturación



```python
ingresos = df.groupby("tienda")["Precio"].sum()
ingresos

```

```python
import matplotlib.pyplot as plt

ingresos.plot(kind="bar")
plt.title("Ingresos totales por tienda")
plt.ylabel("Ingresos ($)")
plt.show()
```

```python
ventas = df.groupby("tienda")["Producto"].count()
ventas
```

# 2. Ventas por categoría

```python
categorias = df.groupby(["tienda", "Categoría del Producto"])["Producto"].count()
categorias
```

```python
categorias = df.groupby(["tienda", "Categoría del Producto"])["Producto"].count().reset_index(name="ventas")

categorias
```

Se analizó la cantidad de productos vendidos por categoría en cada tienda.
En todas las tiendas, la categoría "Muebles" y "Electrónicos" presentan mayor volumen de ventas.
La Tienda 4 mantiene un volumen similar de ventas respecto a las demás, lo que sugiere que su menor ingreso podría deberse a precios promedio más bajos y no necesariamente a menor demanda.

# 3. Calificación promedio de la tienda


```python
calificaciones = df.groupby("tienda")["Calificación"].mean()
calificaciones
```

Se calcularon las calificaciones promedio por tienda.
La Tienda 3 presenta la mejor satisfacción (4.05), mientras que la Tienda 1 muestra la calificación más baja (3.98).
La Tienda 4 tiene una calificación intermedia, por lo que su bajo ingreso no parece estar directamente relacionado con la satisfacción del cliente.

# 4. Productos más y menos vendidos

```python
productos = df.groupby(["tienda", "Producto"])["Producto"].count()
productos
```

```python
productos = productos.reset_index(name="ventas")
productos
```

```python
top_productos = productos.sort_values(["tienda", "ventas"], ascending=[True, False])
top_productos.groupby("tienda").head(5)
```

```python
menos_productos = productos.sort_values(["tienda", "ventas"], ascending=[True, True])
menos_productos.groupby("tienda").head(5)
```

```python
import matplotlib.pyplot as plt

t4 = top_productos[top_productos["tienda"] == "Tienda 4"].head(5)

t4.plot(kind="bar", x="Producto", y="ventas")
plt.title("Top 5 productos más vendidos - Tienda 4")
plt.xticks(rotation=45)
plt.show()
```

```python
top_productos.groupby("tienda").head(5)
```

```python
menos_productos.groupby("tienda").head(5)
```

Se analizaron los productos más y menos vendidos en cada tienda.
Se observó que todas las tiendas presentan volúmenes de ventas similares en sus productos más populares (aprox. 55–65 ventas).
La Tienda 4 no muestra una diferencia significativa en volumen de productos vendidos, lo que sugiere que su menor ingreso no se debe a baja demanda, sino posiblemente a precios promedio más bajos o mayores costos operativos.

# 5. Envío promedio por tienda

```python
envio_promedio = df.groupby("tienda")["Costo de envío"].mean()
envio_promedio
```

```python
import matplotlib.pyplot as plt

envio_promedio.plot(kind="bar")
plt.title("Costo promedio de envío por tienda")
plt.ylabel("Costo envío")
plt.show()
```

```python
precio_promedio = df.groupby("tienda")["Precio"].mean()
precio_promedio
```

Se calculó el costo promedio de envío por tienda.
La Tienda 4 presenta el menor costo de envío, lo que indica que la logística no es el factor que explica su bajo ingreso.
Por el contrario, las Tiendas 1 y 2 presentan costos más elevados.

# Informe

El objetivo de este análisis fue evaluar el desempeño de las cuatro tiendas de la cadena Alura Store para ayudar al Sr. Juan a decidir cuál de ellas debería vender para invertir en un nuevo emprendimiento.

En primer lugar, se calcularon los ingresos totales por tienda. Los resultados mostraron que la Tienda 4 presenta el menor ingreso total, en comparación con las demás tiendas, lo que indica un menor rendimiento financiero general.

Posteriormente, se analizaró la cantidad de productos vendidos por categoría. Se observó que todas las tiendas mantienen volúmenes de ventas similares y comparten categorías populares como muebles y electrónicos, por lo que la Tienda 4 no presenta una baja demanda de productos.

También se evaluaron las calificaciones promedio de los clientes para medir la satisfacción. En este caso, la Tienda 4 mostró calificaciones intermedias y no fue la peor evaluada, lo que sugiere que la experiencia del cliente no es el problema principal.

Luego, se revisaron los productos más y menos vendidos en cada tienda. Nuevamente, la Tienda 4 mostró patrones de ventas similares a las demás, con productos destacados y volúmenes comparables, lo que confirma que mantiene un nivel de ventas competitivo.

Finalmente, se calculó el costo de envío promedio. La Tienda 4 incluso presentó uno de los costos de envío más bajos, por lo que la logística tampoco explica sus menores ingresos.

Al profundizar en el análisis, se calculó el precio promedio de los productos por tienda, donde se identificó que la Tienda 4 tiene el precio promedio más bajo. Esto indica que, aunque vende una cantidad similar de productos, genera menos dinero por cada venta, reduciendo su rentabilidad total.

Con base en todos los análisis realizados, se concluye que la Tienda 4 es la menos eficiente financieramente.
